```python
# initialString = '-b*(-2)+2-√(b↑2*4)-3*4' # false
initialString = '-(x-x)*x+x↑x' # true
# initialString = '-(x*(-√(x+x)+x/x)'
# initialString = '-√(x*x+x*x+x)/(x+x*x)' # true
print(initialString)
```

```python
T = ['+', '⨁', '-', '⊖', '*', '/', '↑', '√', '(', ')']
R = ['C', 'Y', 'M', 'S']
```

```python
parsedString = [el if el in T else 'x' for el in initialString]
print(parsedString)
```

```python
def update(letter, symbol):
    match letter:
        case 'S':
            match symbol:
                case '+':
                    return ['1b', '⨁MYC']
                case '-':
                    return ['1c', '⊖MYC']
                case _:
                    return ['1a', 'MYC']
        case 'C':
            match symbol:
                case '+':
                    return ['2b', '+MYC']
                case '-':
                    return ['2c', '-MYC']
                case _:
                    return ['2a', '']
        case 'Y':
            match symbol:
                case '*':
                    return ['3b', '*MY']
                case '/':
                    return ['3c', '/MY']
                case '↑':
                    return ['3d', '↑MY']
                case _:
                    return ['3a', '']
        case 'M':
            match symbol:
                case '(':
                    return ['4b', '(S)']
                case ')':
                    return ['4b', '(S)']
                case '√':
                    return ['4c', '√(S)']
                case 'x':
                    return ['4A', 'x']
                case _:
                    return ['4a', 'x']
        case _:
            raise Exception('unknown situation')

i = 0
reformedString = 'S'
print(f'reformed string: {reformedString}')

while i < len(initialString):
    letter = None
    letter_ind = None
    symbol = parsedString[i]
    flag = False
    print(f'symbol: {symbol}')
    for ind, el in enumerate(reformedString):
        if symbol == ')' and el == ')':
            flag = True
            break
        if symbol == '(' and el == 'S':
            i += 1
        if el in R:
            letter = el
            letter_ind = ind
            print(f'found letter: {letter}')
            break
    print('out of the loop')
    if flag:
        pass
    else:
        if letter:
            res = update(letter, symbol)
            bridge = res[0]
            replacement = res[1]
            print(f'->{{{bridge}}}')
            reformedString = reformedString[:letter_ind] + replacement + reformedString[(letter_ind+1):]

            if 'a' in bridge:
                i -= 1
        else:
            print('no letters found')
            break

    print(f'reformed string: {reformedString}')
    i += 1
for el in R:
    reformedString = reformedString.replace(el, '')
print(f'result: {reformedString}')
```
